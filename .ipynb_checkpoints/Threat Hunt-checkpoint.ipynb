# Threat Query Class
- A threat query object has the following elements:
    1. nextPage
    2. andQuery
    3. orQuery
    4. notQuery
    5. results
    
- And the following methods:
    1. form_data_query
    2. hunt
    3. filter_hunt

```python
class ThreatQuery:
    def __init__(self):
        self.query = {"query":{"huntQueryVersion":"1.0.0"}}
        self.groupFilter = None
        self.count = None
        self.results = []
        
    
    # Transform the threat hunting URLs based on the groupFilter
    # and count parameters.
    def _transform_url(url):
        if self.groupFilter is not None:
            for group in self.groupFilter:
                url += "&gid={}".format(group)
        if self.count is not None:
            url += "&count={}".format(self.count)
        else:
            url += "&count=20"
        return url
    
    
    def hunt(self, server, url = None):
        # Receive a URL param to switch between inital and more-events hunting.
        if url is None:
            url = self._transform_url("/fapi/query/hunt-v2/events?")
        # Generate the query, re-generated each time as this requires the
        # "nextPage" parameter from the previous hunt.
        data = self._form_data_query()
        reply = api_request(server, url, "POST", data)
        if reply is not None:
            if reply['result'] is not None:
                self.results += reply['result']
                self.nextPage = reply['nextPage']
        else:
            print("Error: Something went wrong. Reply is of None type")
            break
        
    
    # Perform a threat hunt using form_data_query.
    def hunt_multi_page(self, server, page_count):
        # Define the basic API URLs for threat hunting and append
        # group and count modifiers to the API URLs
        base_url = self._transform_url("/fapi/query/hunt-v2/events?")
        more_url = self._transform_url("/fapi/query/hunt-v2/more-events")
        # Loop through the number of pages to hunt for,
        # and hunt until the desired number of pages has been reached.
        for i in range(page_count):
            url = base_url if i == 0 else more_url
            self.hunt(url)

    
    # Perform a threat hunt that will filter results based on
    # the previous result. This allows you to, for example,
    # perform a filter hunt on process created to obtain a list
    # of all unique process created events on a dashboard.
    def filter_hunt(self, server, event_type, filter_key, filter_field)
    
    
    def first_filter_hunt(self, server):
        base_url = self._transform_url("/fapi/query/hunt-v2/events?")
        data = self._form_data_query()
        
        
    
    
    
    def application_hunt(self, server):
        base_url = self._transform_url("/fapi/query/hunt-v2/events?")
        more_url = self._transform_url("/fapi/query/hunt-v2/more-events")
        data = self._form_data_query()
        
        
        
    
    
```

```python
class ThreatQuery:
    def __init__(self):
        self.nextPage = None
        self.andQuery = None
        self.orQuery = None
        self.notQuery = None
        self.groupFilter = None
        self.count = None
        self.results = []
        
    def _set_base_query
        
        
    # Transforms the nextPage, andQuery, orQuery, and notQuery parameters
    # into the threat hunting query.
    def _form_data_query(self):
        if self.nextPage is None:
            query = '{"query":{'
        else:
            query = '{"nextPage":"' + self.nextPage + '","query":{'
        if self.andQuery is not None:
            query += '"andQuery":' + json.dumps(self.andQuery) + ','
        if self.orQuery is not None:
            query += '"orQuery":' + json.dumps(self.orQuery) + ','
        if self.notQuery is not None:
            query += '"notQuery":' + json.dumps(self.notQuery) + ','
        query += '"huntQueryVersion":"1.0.0"}}'
        return query

    
    # Transform the threat hunting URLs based on the groupFilter
    # and count parameters.
    def _transform_url(url):
        if self.groupFilter is not None:
            for group in self.groupFilter:
                url += "&gid={}".format(group)
        if self.count is not None:
            url += "&count={}".format(self.count)
        else:
            url += "&count=20"
        return url
    
    
    def hunt(self, server, url = None):
        # Receive a URL param to switch between inital and more-events hunting.
        if url is None:
            url = self._transform_url("/fapi/query/hunt-v2/events?")
        # Generate the query, re-generated each time as this requires the
        # "nextPage" parameter from the previous hunt.
        data = self._form_data_query()
        reply = api_request(server, url, "POST", data)
        if reply is not None:
            if reply['result'] is not None:
                self.results += reply['result']
                self.nextPage = reply['nextPage']
        else:
            print("Error: Something went wrong. Reply is of None type")
            break
        
    
    # Perform a threat hunt using form_data_query.
    def hunt_multi_page(self, server, page_count):
        # Define the basic API URLs for threat hunting and append
        # group and count modifiers to the API URLs
        base_url = self._transform_url("/fapi/query/hunt-v2/events?")
        more_url = self._transform_url("/fapi/query/hunt-v2/more-events")
        # Loop through the number of pages to hunt for,
        # and hunt until the desired number of pages has been reached.
        for i in range(page_count):
            url = base_url if i == 0 else more_url
            self.hunt(url)

    
    # Perform a threat hunt that will filter results based on
    # the previous result. This allows you to, for example,
    # perform a filter hunt on process created to obtain a list
    # of all unique process created events on a dashboard.
    def filter_hunt(self, server, event_type, filter_key, filter_field)
    
    
    def first_filter_hunt(self, server):
        base_url = self._transform_url("/fapi/query/hunt-v2/events?")
        data = self._form_data_query()
        
        
    
    
    
    def application_hunt(self, server):
        base_url = self._transform_url("/fapi/query/hunt-v2/events?")
        more_url = self._transform_url("/fapi/query/hunt-v2/more-events")
        data = self._form_data_query()
        
        
        
    
    
```

```python
class ProcessProgram:
    def __init__(self):
        self.path = None
        self.filename = None
        self.md5 = None
        self.sha1 = None
        self.sha256 = None
        self.signer = None
        
    def __eq__(self, other):
        if self.path == other.path and self.filename == other.filename and self.md5 == other.md5 and self.sha1 == other.sha1 and self.sha256 == other.sha256 and self.signer == other.signer:
            return True
        else:
            return False
        
    def str(self):
        string = "Path: " + str(self.path) + "\tFilename: " + str(self.filename) + "\nmd5: " + str(self.md5) + "\tsha1: " + str(self.sha1) + "\nsha256: " + str(self.sha256) + "\nsigner: " + str(self.signer)
        return string
    
    def toJSON(self):
        return json.dumps(self, default=lambda o: o.__dict__, sort_keys=True, indent=4)

    
class Signer:
    def __init__(self, signer):
        self.name = signer
        self.programs = None
        
    def __eq__(self, other):
        if self.name == other.name:
            return True
        else:
            return False
        

def construct_filter(programs):
    ## TODO: Add Signers
    md5List = []
    sha1List = []
    sha256List = []
    filenameList = []
    md5Filter = None
    sha1Filter = None
    sha256Filter = None
    filenameFilter = None
    
    for i in programs:
        if i.md5 is not None:
            md5List.append(i.md5)
        elif i.sha1 is not None:
            sha1List.append(i.sha1)
        elif i.sha256 is not None:
            sha256List.append(i.sha256)
        elif i.filename is not None:
            filenameList.append(i.filename)
            
            
    huntFilter = []
    if md5List != []:
        huntFilter.append(["$$maia$$.md5",md5List])
        
    if sha1List != []:
        if huntFilter != "":
            huntFilter += ","
        huntFilter.append(["$$maia$$.sha1",sha1List])
        
    if sha256List != []:
        if huntFilter != "":
            huntFilter += ","
        huntFilter.append(["$$maia$$.sha256",sha256List])
        
    if filenameList != []:
        if huntFilter != "":
            huntFilter += ","
        huntFilter.append(["$$maia$$.filename",filenameList])
        
    return huntFilter
    
def process_created_filter_hunt(server, count):
    pc = ThreatQuery()
    pc.andQuery = []
    pc.andQuery.append(["$$maia$$.eventType",[2]])
    programs = []
    signers = []
    for i in range(count):
        print("Printing " + str(i))
        pc.filter_hunt(server)
        if pc.results is not None:
            for event in pc.results:
                program = ProcessProgram()
                process_data = event['payload']['process']['program']
                if "path" in process_data.keys():
                    program.path = process_data["path"]
                if "filename" in process_data.keys():
                    program.filename = process_data["filename"]
                if "md5" in process_data.keys():
                    program.md5 = process_data["md5"]
                if "sha1" in process_data.keys():
                    program.sha1 = process_data["sha1"]
                if "sha256" in process_data.keys():
                    program.sha256 = process_data["sha256"]
                if "certInfo" in process_data.keys():
                    if "signer" in process_data["certInfo"].keys():
                        program.signer = process_data["certInfo"]["signer"]
                if program not in programs:
                    print("Found new program: " + str(program.filename))
                    programs.append(program)
                    if program.signer is not None:
                        signers.append(program.signer)
                        signer_hunt(server, pc, signers, program.signer, count)
                        pc.andQuery= []
                        pc.andQuery.append(["$$maia$$.eventType",[2]])
        else:
            break
        huntFilter = construct_filter(programs)
        pc.notQuery = []
        for filter in huntFilter:
            pc.notQuery.append(filter)
    return programs

def signer_hunt(server, pc, signers, signer, count):
    pc.andQuery.append([["signer"], [str(signer)]])
    for i in range(count):
        print("Printing " + str(i) + " for " + signer)
        huntFilter = construct_filter(programs)
        pc.notQuery = []
        for filter in huntFilter:
            pc.notQuery.append(filter)
            
        pc.filter_hunt(server)
        if pc.results is not None:
            for event in pc.results:
                program = ProcessProgram()
                process_data = event['payload']['process']['program']
                if "path" in process_data.keys():
                    program.path = process_data["path"]
                if "filename" in process_data.keys():
                    program.filename = process_data["filename"]
                if "md5" in process_data.keys():
                    program.md5 = process_data["md5"]
                if "sha1" in process_data.keys():
                    program.sha1 = process_data["sha1"]
                if "sha256" in process_data.keys():
                    program.sha256 = process_data["sha256"]
                if "certInfo" in process_data.keys():
                    if "signer" in process_data["certInfo"].keys():
                        program.signer = process_data["certInfo"]["signer"]
                if program not in programs:
                    print("Found new program: " + str(program.filename))
                    programs.append(program)
        else:
            break

    
```
